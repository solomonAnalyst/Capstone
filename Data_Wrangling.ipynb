## TASK 1: Calculate the number of launches on each site

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv")

# Apply value_counts() on column 'LaunchSite' to calculate the number of launches per site
launch_counts = df['LaunchSite'].value_counts()

# Display the result
print(launch_counts)
```

## TASK 2: Calculate the number and occurrence of each orbi

```python
# Apply value_counts() on column 'Orbit' to calculate the number and occurrence of each orbit
orbit_counts = df['Orbit'].value_counts()

# Display the result
print(orbit_counts)
```

## TASK 3: Calculate the number and occurence of mission outcome of the orbits

```python
# Apply value_counts() on the 'Outcome' column to calculate the number and occurrence of each mission outcome
landing_outcomes = df['Outcome'].value_counts()

# Display the result
print(landing_outcomes)
```

```python
# Print the index and outcome for each unique landing outcome
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)

# Select bad outcomes using their index positions
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])

# Display bad outcomes
print("Bad outcomes:", bad_outcomes)
```

## TASK 4: Create a landing outcome label from Outcome column

```python
# Create the landing_class list
landing_class = [0 if outcome in bad_outcomes else 1 for outcome in df['Outcome']]

# Assign it as a new column to the DataFrame for better representation
df['LandingClass'] = landing_class

df['Class']=landing_class
df[['Class']].head(8)

```
